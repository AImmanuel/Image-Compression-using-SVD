```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
from skimage import data
from skimage.color import rgb2gray

from skimage import img_as_ubyte,img_as_float
cat_image = rgb2gray(img_as_float(data.chelsea()))  
plt.imshow(cat_image, cmap='gray')
```

```python
from numpy.linalg import svd
```

```python
original_shape = cat_image.shape
print(original_shape) 
```

```python
#   Perform svd decomposition and truncated (using k singular values/vectors) reconstruction
#   returns reconstructed matrix (reconst_matrix), array of singular values (s)

def compress_svd(image,k):

    U,s,V = svd(image,full_matrices=False)
    reconst_matrix = np.dot(U[:,0:k],np.dot(np.diag(s[0:k]),V[0:k,:]))

    return reconst_matrix, s
```

```python
newimg=compress_svd(cat_image,15)

# Compressed image
plt.imshow(newimg[0], cmap='gray')
```
